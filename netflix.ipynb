```
import apache_beam as beam
import csv
p1 = beam.Pipeline()
netflix =(
    p1
    | beam.io.ReadFromText('netflix_titles.csv',skip_header_lines=1)
    | beam.Map(lambda record: record.split(','))
    | beam.Filter(lambda record:(int(record[2])==2020))
    | beam.GroupBy(lambda record:record[1] )
    | beam.io.WriteToText('results.txt')
)
p1.run()
```
